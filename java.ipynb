```python
from matplotlib import pyplot 
y=[]
yy=[]
x=[]
xx=[]
with open ("notes3.txt") as notes:
    line = notes.readline()
       
    list1 = line.split()
     
    for i in list1:
        j=float(i)
        y.append(j)
    print(y)
    line = notes.readline()
    list1 = line.split()
    for i in list1:
        j=float(i)
        x.append(j)
    print(x)
    
    for i in range(len(y)):
        yy.append(i)
    print(yy)
    
    for i in range(len(x)):
        xx.append(i)
    print(xx)
pyplot.errorbar(yy,y,  color='red',label='PercentA')

pyplot.xlabel('Years')
pyplot.ylabel('Percent_A')
pyplot.legend(loc='best')
pyplot.title('PercentA')
pyplot.show()

pyplot.bar(xx,x,  color='red',label='Entropy')

pyplot.xlabel('Agents')
pyplot.ylabel('Entropy')
pyplot.legend(loc='best')
pyplot.title('Entropy')
pyplot.show()
    
        

            
```
